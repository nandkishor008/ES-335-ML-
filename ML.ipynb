ANS-1: To perform a matrix multiplication between a (n, k) and (k, m) matrix.
1.  **Multiplications**:

    

*  For each element in the resulting matrix, you perform K multiplications (since each row in A and column in B has K elements).
*   There are n×m elements in the resultant Matrix.

*   Total multiplications = n×m×k.


2.   **Additions**:

*   After multiplying, you need to add those k products together for each element in the resulting matrix.
*  Adding K numbers requires K-1 additions.

* There are n×m elements in the resultant Matrix.

*   Total Additions = n×m×(k-1).

```python
import time
import numpy as np

# matrices A & B
A = [
    [1, 2, 3],
    [4, 5, 6]
]

B = [
    [7, 8],
    [9, 10],
    [11, 12]
]

# Function to multiply matrices using list of lists to get resultant matrix C.
def matrix_multiply_list(A, B):
    n = len(A)
    k = len(A[0])
    m = len(B[0])
    C = [[0 for _ in range(m)] for _ in range(n)]

    for i in range(n):
        for j in range(m):
            for r in range(k):
                C[i][j] += A[i][r] * B[r][j]

    return C

#time taken for the list of lists method
start_time = time.time()
C_list = matrix_multiply_list(A, B)
end_time = time.time()
list_time = end_time - start_time

print("Result using list of lists:")
for row in C_list:
    print(row)
print(f"Time taken (list of lists): {list_time} sec")

# Convert the lists to numpy arrays
A_np = np.array(A)
B_np = np.array(B)

# time taken for the NumPy method
start_time = time.time()
C_np = np.dot(A_np, B_np)
end_time = time.time()
numpy_time = end_time - start_time

print("\nResult using NumPy:")
print(C_np)
print(f"Time taken (NumPy): {numpy_time} sec")

print("\nTiming Comparison:")
print(f"List of lists time: {list_time} sec")
print(f"NumPy time: {numpy_time} sec")
```

NumPy uses highly optimized libraries like BLAS and LAPACK for numerical computations. therefore solving using numpy is more faster than list of lists.

```python
# Ans 9:

import numpy as np
import matplotlib.pyplot as plt

# domain
x = np.arange(0.5, 100.5, 0.5)

# functions
y1 = x
y2 = x**2
y3 = (x**3) / 100
y4 = np.sin(x)
y5 = np.sin(x) / x
y6 = np.log(x)
y7 = np.exp(x)

# Create the figure and subplots
plt.figure(figsize=(12, 8))

# y=x
plt.subplot(3, 3, 1)
plt.plot(x, y1, label='y=x')
plt.title('y=x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# y=x^2
plt.subplot(3, 3, 2)
plt.plot(x, y2, label='y=x^2', color='orange')
plt.title('y=x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# y=x^3/100
plt.subplot(3, 3, 3)
plt.plot(x, y3, label='y=x^3/100', color='green')
plt.title('y=x^3/100')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# y=sin(x)
plt.subplot(3, 3, 4)
plt.plot(x, y4, label='y=sin(x)', color='red')
plt.title('y=sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# y=sin(x)/x
plt.subplot(3, 3, 5)
plt.plot(x, y5, label='y=sin(x)/x', color='purple')
plt.title('y=sin(x)/x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# y=log(x)
plt.subplot(3, 3, 6)
plt.plot(x, y6, label='y=log(x)', color='brown')
plt.title('y=log(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# y=e^x
plt.subplot(3, 3, 7)
plt.plot(x, y7, label='y=e^x', color='pink')
plt.title('y=e^x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# prevent overlap
plt.tight_layout()
plt.show()
```

```python
#Ans 10:

import numpy as np
import pandas as pd

np.random.seed(0)
matrix = np.random.uniform(1, 2, size=(20, 5))

df = pd.DataFrame(matrix, columns=["a", "b", "c", "d", "e"])

print("DataFrame:\n", df)

std_devs = df.std()
highest_std_dev_column = std_devs.idxmax()
highest_std_dev_value = std_devs.max()


print(f"\nColumn with the Highest Standard Deviation: {highest_std_dev_column} ({highest_std_dev_value})")

row_means = df.mean(axis=1)
lowest_mean_row_index = row_means.idxmin()
lowest_mean_value = row_means.min()

print("\nMean of Each Row:\n", row_means)
print(f"\nRow with the Lowest Mean: Row {lowest_mean_row_index} ({lowest_mean_value})")

```

```python
# Ans 11:

import numpy as np
import pandas as pd

np.random.seed(0)
matrix = np.random.uniform(1, 2, size=(20, 5))

df = pd.DataFrame(matrix, columns=["a", "b", "c", "d", "e"])

df['f'] = df['a'] + df['b'] + df['c'] + df['d'] + df['e']

df['g'] = df['f'].apply(lambda x: 'LT8' if x < 8 else 'GT8')

num_rows_lt8 = df[df['g'] == 'LT8'].shape[0]

std_f_lt8 = df[df['g'] == 'LT8']['f'].std()


std_f_gt8 = df[df['g'] == 'GT8']['f'].std()


print("DataFrame:\n", df)
print(f"\nNumber of rows where the value in column 'g' is 'LT8': {num_rows_lt8}")
print(f"Standard deviation of column 'f' for rows where the value in column 'g' is 'LT8': {std_f_lt8}")
print(f"Standard deviation of column 'f' for rows where the value in column 'g' is 'GT8': {std_f_gt8}")
```

```python
#Ans 13:

import numpy as np

def custom_argmin(arr):

    min_index = 0
    min_value = arr[0]

    for i in range(1, len(arr)):
        if arr[i] < min_value:
            min_value = arr[i]
            min_index = i

    return min_index

# Example
array = np.array([3, 1, 4, 1, 5, 9, 2, 6, 5])

custom_index = custom_argmin(array)
print("Custom argmin index:", custom_index)

numpy_index = np.argmin(array)
print("Numpy argmin index:", numpy_index)
print("Both methods return the same index!")
```
